```python
# Can people accurately predict their own perceived value in the dating market?
# # Select attributes for self vs. others' evaluation
self_evaluated_attributes = ["attr3_1", "sinc3_1", "intel3_1", "fun3_1", "amb3_1"]
peer_evaluated_attributes = ["attr5_1", "sinc5_1", "intel5_1", "fun5_1", "amb5_1"]

# Compute average scores
self_eval = df[self_evaluated_attributes].mean().values
peer_eval = df[peer_evaluated_attributes].mean().values

```

```python
# Organize into a DataFrame
radar_df = pd.DataFrame({
    "Feature": self_evaluated_attributes,  # Labels for the radar axes
    "Self-Evaluation": self_eval,
    "Peer Evaluation": peer_eval
})
radar_df = radar_df.melt(id_vars="Feature", var_name="Evaluation Type", value_name="Score")

# Create Radar Chart
fig = px.line_polar(radar_df, r="Score", theta="Feature", color="Evaluation Type",
                    title="Self-Perception vs. Peer Evaluation in Speed Dating",
                    line_close=True)  # Closing the shape for a web-like effect
fig.show()
```

```python
# Group by gender and compute mean
mean_by_gender = df.groupby("Gender")[self_evaluated_attributes + peer_evaluated_attributes].mean()
```

```python
# Convert male data to long format
male_df = pd.DataFrame({
    "Feature": self_evaluated_attributes, 
    "Self-Evaluation": mean_by_gender.loc["Men", self_evaluated_attributes].values,
    "Peer Evaluation": mean_by_gender.loc["Men", peer_evaluated_attributes].values
}).melt(id_vars="Feature", var_name="Evaluation Type", value_name="Score")

# Convert female data to long format
female_df = pd.DataFrame({
    "Feature": self_evaluated_attributes, 
    "Self-Evaluation": mean_by_gender.loc["Women", self_evaluated_attributes].values,
    "Peer Evaluation": mean_by_gender.loc["Women", peer_evaluated_attributes].values
}).melt(id_vars="Feature", var_name="Evaluation Type", value_name="Score")
```

```python
fig_male = px.line_polar(male_df, r="Score", theta="Feature", color="Evaluation Type",
                         title="Self vs. Peer Evaluation (Men)", line_close=True)

fig_female = px.line_polar(female_df, r="Score", theta="Feature", color="Evaluation Type",
                           title="Self vs. Peer Evaluation (Women)", line_close=True)

fig_male.show()
fig_female.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create the subplot figure with 1 row and 2 columns
fig = make_subplots(rows=1, cols=2, subplot_titles=("Men", "Women"),
                    specs=[[{"type": "polar"}, {"type": "polar"}]])

# Add radar chart for men's self-evaluation (blue)
fig.add_trace(go.Scatterpolar(
    r=male_df.loc[male_df["Evaluation Type"] == "Self-Evaluation", "Score"],
    theta=male_df.loc[male_df["Evaluation Type"] == "Self-Evaluation", "Feature"],
    mode="lines+markers",
    name="Self Evaluation",
    marker=dict(color="blue")
), row=1, col=1)

# Add radar chart for men's peer evaluation (red)
fig.add_trace(go.Scatterpolar(
    r=male_df.loc[male_df["Evaluation Type"] == "Peer Evaluation", "Score"],
    theta=male_df.loc[male_df["Evaluation Type"] == "Peer Evaluation", "Feature"],
    mode="lines+markers",
    name="Peer Evaluation",
    marker=dict(color="red")
), row=1, col=1)

# Add radar chart for women's self-evaluation (blue)
fig.add_trace(go.Scatterpolar(
    r=female_df.loc[female_df["Evaluation Type"] == "Self-Evaluation", "Score"],
    theta=female_df.loc[female_df["Evaluation Type"] == "Self-Evaluation", "Feature"],
    mode="lines+markers",
    name="Self Evaluation",
    marker=dict(color="blue")
), row=1, col=2)

# Add radar chart for women's peer evaluation (red)
fig.add_trace(go.Scatterpolar(
    r=female_df.loc[female_df["Evaluation Type"] == "Peer Evaluation", "Score"],
    theta=female_df.loc[female_df["Evaluation Type"] == "Peer Evaluation", "Feature"],
    mode="lines+markers",
    name="Peer Evaluation",
    marker=dict(color="red")
), row=1, col=2)

# Update layout
fig.update_layout(title_text="Self vs. Peer Evaluation by Gender", showlegend=True, legend_title="Evaluation Type")
fig.show()
```

### Conclusions
